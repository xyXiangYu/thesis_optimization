```python
import numpy as np
import matplotlib.pyplot as plt
def get_optT(p): 
    return 2.5 + 40 /(100 -p ** 2)

def get_optU(p, nr_points=500):
    nr_points = 500
    ts = np.linspace(0, 1, nr_points)
    T = get_optT(p)
    Ts = ts * T
    smaller_bound = 4 /((10-p) * T)
    upper_bound = 1- 4 /((10+p) * T)
    print (T, smaller_bound, upper_bound)


    uts = np.array([p] * nr_points) 
    upper_part = ts >= upper_bound
    smaller_part = ts < smaller_bound
    uts[upper_part] = -10
    uts[smaller_part] = 10

    return T, uts, Ts, smaller_bound, upper_bound
```

```python
p0 = 0 
T0, uts0, Ts0, Ts0_small, Ts0_up = get_optU(p0)
p5 = 5 
T5, uts5, Ts5, Ts5_small, Ts5_up = get_optU(p5)
p9 = 9 
T9, uts9, Ts9, Ts9_small, Ts9_up = get_optU(p9)
```

```python
# plot results
plt.figure(1)
plt.subplot(3,1,1)
plt.plot(Ts0,uts0,'r-',lw=2)
plt.ylabel('u(t)')
plt.legend(['u(t) when p=0'])

plt.subplot(3,1,2)
plt.plot(Ts5,uts5,'b-',lw=2)
plt.ylabel('u(t)')
plt.legend(['u(t) when p=5'])

plt.subplot(3,1,3)
plt.plot(Ts9, uts9,'g-',lw=2)
plt.ylabel('u(t)')
plt.legend(['u(t) when p=9'])

plt.xlabel('Time')
plt.title(f"theoretical solution when p=0, 5 and 9, with T={T0}, {round(T5,4)}, {round(T9,4)} respectively", y=-0.6)
plt.show()
```
